```python
#MODEL 2799
datat = pd.read_csv('Ptensor_Data_MODEL_2799_ics_6.0.csv')
df1 = pd.DataFrame(datat, columns = ['wavenumber k', 'h h N=67.6'])

dataz = pd.read_csv('Pzeta_Data_MODEL_2799_ics_6.0.csv')
df2 = pd.DataFrame(dataz, columns = ['wavenumber k', 'cal(P)_zeta N=67.6'])

df3 = pd.DataFrame(datat, columns = ['wavenumber k', 'nt_h h N=67.6'])

rescale = 3.65124853e-02

x1 = (df1['wavenumber k'])*rescale
y1 = df1['h h N=67.6']
x10 = [0.002,0.002]   #plots vertical line for x=0.002
y10 = [y1[0],y1[100]]
x100 = []
y100 = []
y0 = df2['cal(P)_zeta N=67.6']

r = y1[5]/y0[5]


def power_law(x, a, b):
    return a+b*x
def power_law2(x, a, b):
    return np.exp(a)*np.power(x,b)

#loop for getting the required points for fitting
q=0
while q !=100:
    x100.append(x1[q]);
    y100.append(y1[q]);
    q = q+1;

x100_log = np.log(x100)
y100_log = np.log(y100)


#power law fitting
pars, cov = curve_fit(f=power_law, xdata=x100_log, ydata=y100_log, p0=[0, 0], bounds=(-np.inf, np.inf))
p1 = power_law2(x100, pars[0], pars[1])

print('amplitude: ',pars[0],'\nr: ',r*4,'\nrBigRunData: ',rtrue[2798],'\nntBigRunData: ',nt[2798],'\ntensor spectral index: ', pars[1],'\ninternal estimate of nt: ',df3['nt_h h N=67.6'][5])
print('Consistency Bound: ',r < -8*pars[1])

pl.loglog(x1, y1, label = 'Model 2799')
pl.plot(x10,y10,label = 'k = 0.002')

pl.plot(x100, p1, label = 'Fit')

pl.legend()
pl.xlabel('log (wavenumber k) mpc^-1')
pl.ylabel('h h N=63.2')

pl.show()
```
